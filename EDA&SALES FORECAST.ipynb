Sales Over Time

```python
import matplotlib.pyplot as plt

df_sales['ORDERDATE'] = pd.to_datetime(df_sales['ORDERDATE'])

daily_sales = df_sales.groupby(df_sales['ORDERDATE'].dt.date)['ORDERQUANTITY'].sum().reset_index()

# Plot sales over time
plt.figure(figsize=(12, 6))
plt.plot(daily_sales['ORDERDATE'], daily_sales['ORDERQUANTITY'], marker='o')
plt.title('Daily Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

Monthly Sales Performance

```python
df_sales['month'] = df_sales['ORDERDATE'].dt.to_period('M')
monthly_sales = df_sales.groupby('month')['ORDERQUANTITY'].sum().reset_index()

plt.figure(figsize=(12, 6))
monthly_sales['month'] = monthly_sales['month'].astype(str)
plt.bar(monthly_sales['month'], monthly_sales['ORDERQUANTITY'])
plt.title('Monthly Sales Performance')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.ylabel('Total Sales')
plt.show()
```

Top-Performing Products

```python
product_sales = df_sales.groupby('PRODUCTKEY')['ORDERQUANTITY'].sum().reset_index()

# Identify top 10 products
top_products = product_sales.sort_values(by='ORDERQUANTITY', ascending=False).head(10)

print(top_products)
```

Sales by Category

```python
sales_products = df_sales.merge(df_prod, on='PRODUCTKEY')
sales_subcategories = sales_products.merge(df_sub_cat, on='PRODUCTSUBCATEGORYKEY')
sales_categories = sales_subcategories.merge(df_cat, on='PRODUCTCATEGORYKEY')

sales_by_category = sales_categories.groupby('CATEGORYNAME')['ORDERQUANTITY'].sum().reset_index()

# Plot sales by category
plt.figure(figsize=(10, 6))
plt.bar(sales_by_category['CATEGORYNAME'], sales_by_category['ORDERQUANTITY'], color='skyblue')
plt.title('Sales by Product Category (Order Quantity)')
plt.xlabel('Category')
plt.ylabel('Total Order Quantity')
plt.xticks(rotation=45)
plt.show()
```

# forecast future sales trends

```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# Prepare sales data for time series analysis
df_sales['ORDERDATE'] = pd.to_datetime(df_sales['ORDERDATE'], errors='coerce')
df_sales = df_sales.dropna(subset=['ORDERDATE'])  # Drop rows with invalid dates
df_sales = df_sales.set_index('ORDERDATE').sort_index()

# Aggregate sales by date
daily_sales = df_sales['ORDERQUANTITY'].resample('D').sum()

# Fill missing dates with zero sales
daily_sales = daily_sales.fillna(0)

# Time series decomposition
decomposition = seasonal_decompose(daily_sales, model='additive', period=365)
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

# Plot decomposition
plt.figure(figsize=(14, 8))
plt.subplot(411)
plt.plot(daily_sales, label='Original')
plt.legend(loc='upper left')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='upper left')
plt.subplot(413)
plt.plot(seasonal, label='Seasonality')
plt.legend(loc='upper left')
plt.subplot(414)
plt.plot(residual, label='Residuals')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()

# Fit ARIMA model
model = ARIMA(daily_sales, order=(5, 1, 0))
model_fit = model.fit()

# Forecast
forecast_steps = 365  # Forecast for the next year
forecast = model_fit.forecast(steps=forecast_steps)
forecast_index = pd.date_range(start=daily_sales.index[-1] + pd.Timedelta(days=1), periods=forecast_steps, freq='D')
forecast_series = pd.Series(forecast, index=forecast_index)

# Plot forecast
plt.figure(figsize=(12, 6))
plt.plot(daily_sales, label='Historical Sales')
plt.plot(forecast_series, label='Forecasted Sales', color='red')
plt.title('Sales Forecast')
plt.xlabel('Date')
plt.ylabel('Order Quantity')
plt.legend()
plt.grid(True)
plt.show()
```
